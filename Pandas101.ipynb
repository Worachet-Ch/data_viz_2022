HW (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)

1.   คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ของข้อมูลทั้งหมด
2.   คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ของผู้ป่วยในจังหวัดขอนแก่น
3. หาจำนวนผู้ป่วยที่เป็นคน "**ไม่ใช่คนไทย**"

````
this_data = data_covid[['sex','age','province_of_onset']]
````

````
this_data
````

````

men = this_data[this_data['sex']=='ชาย']
men['age']
````

````
#ข้อ 1
sum = 0
N = 0
for a in men['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย {sum/N}')
````

````
#ข้อ 2
men_KK = men[men['province_of_onset']=='ขอนแก่น']

sum = 0
N = 0
for a in men_KK['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ยของผู้ป่วยชายในจังหวัดขอนแก่น {sum/N}')
````

````
#ข้อ 3
foreigner =  data_covid[data_covid['nationality'].notnull()] #สร้างตัวแปร foreigner รับค่า data_covid คอลัมน์ nationality โดยไม่เอาค่า NaN
foreigner_quiz = foreigner[foreigner['nationality']!='Thailand'] #สร้างตัวแปร foreigner_quiz รับค่า foreigner คอลัมน์ nationality โดยไม่เอา Thailand
num = foreigner_quiz['No.'] 
foreigner_quiz[['No.','nationality']] #แสดงคอลัมน์ No and nationality
````

````
#ข้อ 3
sum = 0
for a in num:
  if a>0:
    sum +=1
print(f'ผู้ป่วยที่ไม่ใช่คนไทยทั้งหมด {sum} คน')
````

### การวนลูป record ในตาราง  (```.iterrows```)

````
this_data = data_covid[['sex','age','province_of_onset']]
````

````
for each_row in this_data.iterrows():
    # print(each_row)
    # print(each_row[1])
    if (each_row[1]['age'] == 20) and (each_row[1]['province_of_onset'] == 'ขอนแก่น'):
        print(each_row)
````

### การวนลูป แบบมองตารางแพนด้าส์(pandas dataframe) เป็น numpy array  (```.iloc```)

````
for each_row in range(this_data.shape[0]):
    if (this_data.iloc[each_row,1] == 20) and (this_data.iloc[each_row,2] == 'ขอนแก่น'):  #each_row คือ ตัวเลข
        print(each_row)
        print(this_data.iloc[each_row,:])   
````
