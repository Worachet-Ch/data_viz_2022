## Dropna

## แก้คำผิด

```python
alldb_dp['user'].replace({'45-161':'45161','745-61':'74561','71378\n71378':'71378','40591\n40591\n':'40591','17536\n\n17536':'17536',},inplace=True)
```

## ต้องการให้ RowKey, phone และ urine เป็น String

```python
alldb_dp['phone']=alldb_dp['phone'].astype('string')
alldb_dp['RowKey']=alldb_dp['RowKey'].astype('string')
alldb_dp['urine']=alldb_dp['urine'].astype('string')
```

## Formate ให้เป็นวันที่

```python
alldb_dp['date'] = pd.to_datetime(alldb_dp['date'],format="%Y-%m-%d") 
alldb_dp
```

## Formate ให้เป็นเวลา

```python
alldb_dp['time'] = pd.to_datetime(alldb_dp['time'],format='%H:%M%:%S',errors='ignore')
alldb_dp
```

## เปลี่ยนรูปแบบจาก : เป็น -

```python
alldb_dp['time'] = alldb_dp['time'].replace(':','-', regex=True)
```

```python
set(alldb_dp['time'])
```

## ทำให้คอลัมน์ user, site, ov และ PartitionKey มี 5 หลัก

```python
US=list() # สร้าง list ว่าง
for each_row in alldb_dp.iterrows(): #.iterrows เพื่อให้เราสามารถ Loop และเข้าถึงข้อมูลทีละ Row ได้เลย ผ่าน Function ที่ชื่อว่า Iterrows
    if len(str(each_row[1]['user']))== 5 : # len(str เอา str ที่มี 5 หลัก
       if str(each_row[1]['user']) <= '99999' : # เงื่อนไขเอาที่มีค่าน้อยกว่าเท่ากับ 99999
          US.append(True)  # ถ้าตรงเงื่อนไขก็จะเอาที่ตรงเงื่อนไขมาต่อกัน
       else: 
          US.append(False)  
    else: 
          US.append(False) # ถ้าไม่ตรงเงื่อนไขก็จะเอาออกไป
alldb_dp = alldb_dp[US]
alldb_dp
```

```python
SI=list() 
for each_row in alldb_dp.iterrows(): 
    if len(str(each_row[1]['site']))== 5 : 
        if str(each_row[1]['site']) <= '99999' : 
          SI.append(True) 
        else: 
          SI.append(False)  
    else: 
          SI.append(False) 
alldb_dp = alldb_dp[SI]
alldb_dp
```

```python
PK=list() 
for each_row in alldb_dp.iterrows(): 
    if len(str(each_row[1]['PartitionKey']))== 5 : 
        if str(each_row[1]['PartitionKey']) <= '99999' : 
          PK.append(True) 
        else: 
          PK.append(False)  
    else: 
          PK.append(False) 
alldb_dp = alldb_dp[PK]
alldb_dp
```

## แทนข้อมูล fillna()

```python
hospital_c = hospital.fillna(value={'name':'ไม่ระบุ','zone_name':'ไม่ระบุ','amphur':'ไม่ระบุ','tambon':'ไม่ระบุ','name5':'ไม่ระบุ','name10':'ไม่ระบุ','short_thai':'ไม่ระบุ','short_english':'ไม่ระบุ'})
hospital_c
```

## แทนข้อมูลด้วย mean

```python
pc = hospital_c[['provincecode']].mean() 
hospital_c[['provincecode']] = hospital_c[['provincecode']].fillna(pc)
```

```python
ac = hospital_c[['amphurcode']].mean() 
hospital_c[['amphurcode']] = hospital_c[['amphurcode']].fillna(ac)
```

```python
tc = hospital_c[['tamboncode']].mean() 
hospital_c[['tamboncode']] = hospital_c[['tamboncode']].fillna(tc)
```

```python
c4 = hospital_c[['code4']].mean() 
hospital_c[['code4']] = hospital_c[['code4']].fillna(c4)
```

```python
c9 = hospital_c[['code9']].mean() 
hospital_c[['code9']] = hospital_c[['code9']].fillna(c9)
```

```python
hospital_c.isnull().any()
```

```python
hospital_c1 = hospital_c [hospital_c['hcode']!= '2475x']
```

```python
hospital_c2 = hospital_c1 [hospital_c1['hcode']!= '2560x']
```

## hcode แค่ 5 หลักเพื่อเชื่อมตารางกับ alldb

```python
HS=list() 
for each_row in hospital_c2.iterrows(): 
    if len(str(each_row[1]['hcode']))== 5 : 
        if str(each_row[1]['hcode']) <= '99999' : 
          HS.append(True) 
        else: 
          HS.append(False)  
    else: 
          HS.append(False) 
hospital_c2 = hospital_c2[HS]
hospital_c2
```

```python
hospital_c2
```

# รวมข้อมูล alldb กับ hospital ทั้งสองตารางเข้าด้วยกัน

```python
alldb_hos = alldb_dp.merge(hospital_c2,how='left',left_on='site',right_on='hcode') 
alldb_hos
```

```python
alldb_hos.isnull().any()
```

```python
alldb_hosdp = alldb_hos.dropna()
alldb_hosdp
```
